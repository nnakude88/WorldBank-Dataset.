```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
wb = pd.read_excel("WorldBank.xlsx")
wb
```

```python
wb.columns
```

```python
wb.shape
```

```python
wb.dtypes
```

```python
wb.isnull().sum()
```

```python
wb["Birth rate, crude (per 1,000 people)"].mean()
```

```python
wb["Birth rate, crude (per 1,000 people)"] = wb["Birth rate, crude (per 1,000 people)"].fillna(28)
```

```python
wb["Death rate, crude (per 1,000 people)"].mean()
```

```python
wb["Death rate, crude (per 1,000 people)"] = wb["Death rate, crude (per 1,000 people)"].fillna(10.5)
```

```python
wb["Electric power consumption (kWh per capita)"].mean()
```

```python
wb["Electric power consumption (kWh per capita)"] = wb["Electric power consumption (kWh per capita)"].fillna(3175)
```

```python
wb["GDP (USD)"].mode()
```

```python
wb["GDP (USD)"] = wb["GDP (USD)"].fillna(1.150000e+11)
```

```python
wb["GDP per capita (USD)"].mean()
```

```python
wb["GDP per capita (USD)"] = wb["GDP per capita (USD)"].fillna(8231)
```

```python
wb["Individuals using the Internet (% of population)"].mean()
```

```python
wb["Individuals using the Internet (% of population)"] = wb["Individuals using the Internet (% of population)"].fillna(23)
```

```python
wb["Infant mortality rate (per 1,000 live births)"].mean()
```

```python
wb["Infant mortality rate (per 1,000 live births)"] = wb["Infant mortality rate (per 1,000 live births)"].fillna(51)
```

```python
wb["Life expectancy at birth (years)"].mean()
```

```python
wb["Life expectancy at birth (years)"] = wb["Life expectancy at birth (years)"].fillna(64)
```

```python
wb["Population density (people per sq. km of land area)"].mean()
```

```python
wb["Population density (people per sq. km of land area)"] = wb["Population density (people per sq. km of land area)"].fillna(318)
```

```python
wb["Unemployment (% of total labor force) (modeled ILO estimate)"].mode()
```

```python
wb["Unemployment (% of total labor force) (modeled ILO estimate)"] = wb["Unemployment (% of total labor force) (modeled ILO estimate)"].fillna(3.6)
```

```python
wb.isnull().sum()
```

```python
wb["Region"].unique()
```

```python
wb["Year"].unique()
```

```python
wb["population_growth_rate"] =wb["Birth rate, crude (per 1,000 people)"] - wb["Death rate, crude (per 1,000 people)"] / 10 
```

```python
wb["Total_electricity_consumption"] = wb["Death rate, crude (per 1,000 people)"] * wb["Population density (people per sq. km of land area)"]
```

```python
wb["Total_GDP"] = wb["GDP per capita (USD)"] * wb["Population density (people per sq. km of land area)"]
```

```python
wb.columns
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP per capita (USD)', y='Life expectancy at birth (years)', data=wb)
plt.title('GDP per capita vs. Life expectancy at birth')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life expectancy at birth (years)')
plt.grid(True)
plt.show()
```

```python
# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='GDP (USD)', data=wb)
plt.title('GDP over years')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.grid(True)
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='IncomeGroup', y='Infant mortality rate (per 1,000 live births)', data=wb)
plt.title('Infant mortality rate by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Infant mortality rate (per 1,000 live births)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Scatter plot for GDP per capita vs. Internet usage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP per capita (USD)', y='Individuals using the Internet (% of population)', data=wb)
plt.title('GDP per capita vs. Internet Usage')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Internet Usage (% of population)')
plt.grid(True)
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = wb[['GDP per capita (USD)', 'Life expectancy at birth (years)', 'Infant mortality rate (per 1,000 live births)',
                  'Population density (people per sq. km of land area)', 'Unemployment (% of total labor force) (modeled ILO estimate)']].corr()
# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap')
plt.show()
```
